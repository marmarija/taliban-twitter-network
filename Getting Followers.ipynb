```python
import tweepy
import json
import requests
import pandas as pd 
import time

consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""


client = tweepy.Client(consumer_key= consumer_key,
                       consumer_secret= consumer_secret,
                       access_token= access_token,
                       access_token_secret= access_token_secret)


# Twitter API V2 documentation
# Followers for specific ID:
# GET /2/users/:id/followers

# ID Following:
# GET /2/users/:id/following

# https://developer.twitter.com/en/docs/twitter-api/users/follows/api-reference/get-users-id-following

# Tweepy documentation:
# https://docs.tweepy.org/en/stable/client.html


# Example - Id of @anashaqqani313

check_id = '1289055510117748741'

# User_auth = True, for sending API request

username = []
user_id = []
name = []
created_at = []
followers_count = []
following_count = []
tweet_count = []


# Pagination needed in case ID has more than 100 followers
# Pagination documentation:
# https://docs.tweepy.org/en/latest/v2_pagination.html
# User Fields - additional metrics

paginator = tweepy.Paginator(
    client.get_users_following,              
    id = check_id,       
    max_results=100, 
    user_auth = True,
    user_fields=["username","id","created_at","location","public_metrics"]
).flatten(limit=310000)

# For getting followers:
# client.get_users_followers
# Limit 310000, because Haqqani has over 310k followers

results = []
counter = 0

# Adding counter for setting time sleep and checking if the loop breaks

for page in paginator:
    try:
        results.append(page.data)
        for follow in results:
            select_count = int(follow['public_metrics']['followers_count'])
            counter +=1
            if select_count > 1000:
                print(follow['username'])
                print(follow['id'])
                print(follow['name'])
                print(follow['public_metrics']['followers_count'])
                
                # Only IDs with more than 1000 followers
                username.append(follow['username'])
                user_id.append(follow['id'])
                name.append(follow['name'])
                created_at.append(follow['created_at'])
                followers_count.append(follow['public_metrics']['followers_count'])
                following_count.append(follow['public_metrics']['following_count'])
                tweet_count.append(follow['public_metrics']['tweet_count'])
            
            # Idea for time sleep:
            # https://stackoverflow.com/questions/43447678/i-used-time-sleep-but-twittertweepy-crawling-error-429-python
            
            if counter%40 == 0:
                time.sleep(3.3) 
            elif counter%50 == 0:
                time.sleep(8)
            elif counter%60 == 0:
                time.sleep(7)
            elif counter%70 == 0:
                time.sleep(6) 
            elif counter%80 == 0:
                time.sleep(4)
            elif counter%90 == 0:
                time.sleep(3)    
            elif counter%100 == 0:
                time.sleep(9)
                
                # Saving data for every 100 results, in case the loop breaks
                
                df = pd.DataFrame({"Username" : username, 
                   "User ID": user_id,
                   "Name" : name,
                  "Created At": created_at,
                   "Followers Count": following_count,
                   "Following Count": following_count,
                   "Tweet Count": tweet_count
                  
                  })
                df.to_csv("test111.csv", index=False)
   
            else:
                time.sleep(0.33)
            print(counter)    
        
        # Emptying the list after each iteration, in order to avoid duplicating data 
        results=[]
    
    except StopIteration:
        break  

df = pd.DataFrame({"Username" : username, 
                   "User ID": user_id,
                   "Name" : name,
                  "Created At": created_at,
                   "Followers Count": following_count,
                   "Following Count": following_count,
                   "Tweet Count": tweet_count
                  
                  })
df.to_csv("test.csv", index=False)
    
```
